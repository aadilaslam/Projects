```python
A = float(1.85E-12) #Arrhenius Parameter
print(A)
Ea = 14.1*1000 #Activation energy in J mol^-1
print(Ea)
R = 8.314 #Molar gas constant in J mol^-1 K^-1
T_10a = (temp_4d_array + 1.5) #All possible temperatures increased by 1.5 K

k_10a = A*np.exp(-Ea/(R*T_10a)) #New rate coefficents
print(k_10a)
```

```python
global_methane_sink_10a = np.mean(k_10a[:,:,:,:38]*oh_array)
print(global_methane_sink_10a)
```

```python
lifetime_10a = np.reciprocal(global_methane_sink_10a) #The lifetime is the reciprocal of the methane sink
print(lifetime_10a, 'seconds')

time_in_years = (lifetime_10a*(3.171E-8))
print(time_in_years, 'years')
```

```python
delta_lifetimes = ((lifetime - lifetime_10a)*(3.171E-8))
print(round(delta_lifetimes, 2)) #Difference in years rounded to 2 decimal places.
```

The change in temperature by 1.5 K extends the annual mean lifetime of methane by 0.32 years.<br>This can be represented graphically like we did in Question 4.

```python
methane_ratio_10a=np.mean(k_10a[:,:,:,:38]*oh_array, axis=3) #average over months but not latitude or longitude

methane_ratio_10a_2=np.mean(methane_ratio_10a, axis=0) # we need to transpose the data to get it aligned with the earth
# plot on the earth background

methane_ratio_10a_2d=np.transpose(methane_ratio_10a_2)

methane_lifetime_10a = np.reciprocal(methane_ratio_10a_2d)

methane_lifetime_10a_years = (methane_lifetime_10a*(3.171E-8))

fig = plt.figure(figsize=[10,6])
ax = plt.axes(projection=ccrs.PlateCarree()) #projection determines how we view the globe in 2D
ax.coastlines() # add coastlines
ax.set_xticks([-180,-90, 0, 90,180], crs=ccrs.PlateCarree())
ax.set_yticks([-60, -30, 0, 30, 60], crs=ccrs.PlateCarree())
plt.contourf(lon_array,lat_array,methane_lifetime_10a_years,25,cmap=plt.cm.get_cmap('Spectral_r'))
plt.xlabel('Longitude')
plt.ylabel('Latitude')
cbar=plt.colorbar(orientation='horizontal', shrink=0.8)
cbar.set_label('Years')
plt.title('Methane lifetime by an increase of 1.5 K')
plt.show()
```
