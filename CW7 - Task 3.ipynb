```python
nums=1000     
fract=0.01   
recovr= 0.2         
num_infect=1      
numd=40                                
reps=40                             
data= np.arange(numd)               
sdata= np.arange(numd)  
remov=0                                

for k in range (reps):     
    P=np.arange(nums)                 
    for i in range (nums):                
        P[i]=1                                
    fn=np.trunc(fract*nums)               
    n=0
    while n < fn:                       
        ra=np.random.randint(nums)
        if P[ra] != 0:
            P[ra]=0
            n=n+1
    n=0
    while n < 1:                        
        ra = np.random.randint(nums)
        if P[ra]==1:
            P[ra]=2
            n=1

    for j in range (numd):               
        for i in range (nums):             
            if P[i] == 2:                      
                for ii in range (num_infect):
                    ra=np.random.randint(nums)
                    if P[ra] == 1:
                        P[ra] = 2
                        remov=remov+1                         
            if np.random.rand()< recovr: 
                ra=np.random.randint(nums)
                if P[ra] == 2:
                    P[ra]=0 
        c=0
        s=0
        for i in range (nums):               
            if P[i] == 2:
                c=c+1
            if P[i] == 1:
                s=s+1
        data[j]=data[j]+c                   
        sdata[j]=sdata[j]+s                   

num_infected=remov*1.0/reps
print("Average number of students infected =",num_infected)

x=np.arange(numd)
y1=sdata/reps # Number of Susceptible Students
y2=data/reps # Number of Infected Students 
y3=nums-fn-(sdata/numd) # Number of Affected Students
plt.plot(x,y1,'g',x,y2,'r',x,y3,'k')
plt.xlim(0,40)
plt.ylim(0,1000)
plt.title("A plot showing the spread of infection over 40 days")
plt.xlabel("Number of days")
plt.ylabel("Number of students")
plt.legend(['Susceptible','Infected','Affected'])
plt.show() 
```

```python
nums=1000     
fract=0.01   
recovr= 0.2         
num_infect=1      
numd=40                                
reps=40                            
data= np.arange(numd)               
sdata= np.arange(numd)  
remov=0                                

for k in range (reps):     
    P=np.arange(nums)                 
    for i in range (nums):                
        P[i]=1                                
    fn=np.trunc(fract*nums)               
    n=0
    while n < fn:                       
        ra=np.random.randint(nums)
        if P[ra] != 0:
            P[ra]=0
            n=n+1
    n=0
    while n < 1:                        
        ra = np.random.randint(nums)
        if P[ra]==1:
            P[ra]=2
            n=1

    for j in range (numd):               
        for i in range (nums):             
            if P[i] == 2:                      
                for ii in range (num_infect):
                    ra=np.random.randint(nums)
                    if P[ra] == 1:
                        P[ra] = 2
                        remov=remov+1                         
            if np.random.rand()< recovr: 
                ra=np.random.randint(nums)
                if P[ra] == 2:
                    P[ra]=1 
        c=0
        s=0
        for i in range (nums):               
            if P[i] == 2:
                c=c+1
            if P[i] == 1:
                s=s+1
        data[j]=data[j]+c                   
        sdata[j]=sdata[j]+s                   

num_infected=remov*1.0/reps

x=np.arange(numd)
y1=sdata/reps # Number of Susceptible Students
y2=data/reps # Number of Infected Students 
y3=nums-fn-(sdata/numd) # Number of Affected Students 
plt.plot(x,y1,'g',x,y2,'r',x,y3,'k--')
plt.xlim(0,40)
plt.ylim(0,1000)
plt.title("Plot to show the spread of disease over 40 days")
plt.xlabel("Number of days")
plt.ylabel("Number of students")
plt.legend(['Susceptible','Infected','Affected'])
plt.show() 
```

After a certain time period an equilibrium is established between the number of susceptible students and the affected & infected students. When someone recovers from the disease, they have a chance of being re-infected and thereby cannot gain immunity. This is classed as an endemic as the duration of the disease has no end date.
